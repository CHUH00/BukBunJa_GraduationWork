## 전처리

### 1회차 ~ 600회차

```python
column = [
    '년도', '회차', '추첨일', '1등_당첨자수', '1등_당첨금액',
    '2등_당첨자수', '2등_당첨금액', '3등_당첨자수', '3등_당첨금액', 
    '4등_당첨자수', '4등_당첨금액', '5등_당첨자수', '5등_당첨금액', 
    '당첨번호_1', '당첨번호_2', '당첨번호_3', '당첨번호_4', '당첨번호_5', '당첨번호_6', '보너스번호'
]

lotto600_df.columns = column
lotto600_df = lotto600_df.drop(index=[0, 1], axis=1).reset_index(drop=True)
lotto600_df.head()
```

```python
lotto600_df.info()
```

```python
# 결측치 처리
for i in range(600):
    lotto600_df.loc[i, '년도'] = lotto600_df.loc[i, '추첨일'][0:4]
    
# 추첨일 정수 변환
lotto600_df['추첨일'] = lotto600_df['추첨일'].str.replace('.', '')

# , and 원 제거
for i in range(1, 6):
    lotto600_df[f'{ i }등_당첨금액'] = (
        lotto600_df[f'{ i }등_당첨금액']
        .str.replace(',', '')
        .str.replace('원', '')
    )
    
lotto600_df = lotto600_df.astype(int)

lotto600_df.info()
lotto600_df.head()
```

### 601회차 ~ 1184회차

```python
column = [
    '년도', '회차', '추첨일', '1등_당첨자수', '1등_당첨금액',
    '2등_당첨자수', '2등_당첨금액', '3등_당첨자수', '3등_당첨금액', 
    '4등_당첨자수', '4등_당첨금액', '5등_당첨자수', '5등_당첨금액', 
    '당첨번호_1', '당첨번호_2', '당첨번호_3', '당첨번호_4', '당첨번호_5', '당첨번호_6', '보너스번호'
]

lotto1184_df.columns = column
lotto1184_df = lotto1184_df.drop(index=[0, 1], axis=1).reset_index(drop=True)
lotto1184_df.head()
```

```python
lotto1184_df.info()
```
